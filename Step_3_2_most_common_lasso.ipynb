# Feature group frequencies

```python
### GROUPS

## Lexical sophistication
LS = ['nr_stopwords', 'avg_sentence_len_words', 'unique_words', 'word_count', 'char_count', 'avg_word_len', 'nr_long_words', 'nr_short_words',
'most_freq_word_length', 'nr_sentences', 'nr_long_sentences','nr_short_sentences', 'most_freq_sentence_length']

## Readability measures
RM = ['dale_chall_readability', 'gunning_fox_index', 'flesh_reading_ease', 'flesh_kincaid_grade_level','automated_readability_index', 'SMOG', 'LIX', 'OVIX', 'nominal_ratio']

## Lexical diversity
LD = ['TTR', 'guirauds_index', 'yules_k', 'd_estimate', 'hapax_legomena', 'advanced_guiraud']

## Grammar
G = ['nr_unique_pos', 'avg_tree_height', 'nr_grammar_errors', "correct_verb_form"]

## POS tag
POS = ['comparative_adj', 'superlative_adj', 'modal_aux', 'participle', 'infinitive_marker', 'verb_baseform', 'verb_past_tense', 
# coordinating conjunction
'CCONJ', 
# subordinating conjunction
"SCONJ", 
# numeral
'NUM',
# determiner
'DET', 
# preposition
'ADP',
# adjektiv
'ADJ', 
# adverb 
'ADV', 
# pronoun
"PRON",
# proper noun
'PROPN', 
# common noun
'NOUN']

## Mechanics
M = ['nr_spelling_errors', 'nr_capitalization_errors', 'nr_punctuation_errors']

## Content
C = ['grade_as_feature','avg_cosine_similariy_high_grade', 'pattern_cosine', 'weighted_cosine']

#### Coherence ####

## Basic coherence measures
# Average distance between neighbouring points 
BC = ['avg_neighbouring_eucl_dist', 'avg_neighbouring_cos_similarity', 
# Minimum distance between neighbouring points 
'min_neighbouring_eucl_dist','min_neighbouring_cos_similarity',
# Maximum distance between neighbouring points 
'max_neighbouring_eucl_dist', 'max_neighbouring_cos_similarity', 
# Index (minimum distance/maximum distance) 
'eucl_min_max_quotient','cos_min_max_quotient',
# Average distance between any two points
"avg_eucl_dist", "avg_cos_similarity",
# Maximum difference between any two points
"max_eucl_dist", "max_cos_similarity",
# Clark’s and Evans’ distance to nearest neighbour
'eucl_clark_evan_dist_nn', # 'cos_clark_evan_dist_nn'
# Average distance to nearest neighbour
"avg_eucl_dist_nn", #"avg_cos_similarity_nn",
# Cumulative frequency distribution
'avg_eucl_cumulative_freq_distribution']#'avg_cos_cumulative_freq_distribution',  

## spatial data analysis
# Average distance between points and centroid 
SDA = ['avg_eucl_centroid_distances','avg_cos_centroid_similarities', 
# Minimum distance between points and centroid
'min_eucl_centroid_distances','min_cos_centroid_similarities',
# Maximum distance between points and centroid 
'max_eucl_centroid_distances','max_cos_centroid_similarities', 
# Index (minimum distance/maximum distance) 
'eucl_centroid_min_max', 'cos_centroid_min_max',
# Standard distance
'standard_distance',
# Relative distance
'eucl_relative_distance',#'cos_relative_distance', 
# Determinant of distance matrix
'det_eucl_dist_matrix'] #'det_cos_similarity_matrix'

## spatial autocorrelation
SA = ['morans_i', 'gearys_c', 'gettis_g']
```

```python
feature_groups = {"LS":LS, "RM": RM, "LD": LD, "G":G, "POS":POS, "M":M, "C":C, "BC":BC, "SDA":SDA, "SA":SA}
feature_groups
```

```python
def get_groups(df, DS):

    DS1_df = pd.DataFrame(df[DS]).copy()
    DS1_df["feature_group_LS"] = DS1_df[DS].apply(lambda feature: "LS" if feature in feature_groups["LS"] else 0)
    DS1_df["feature_group_RM"] = DS1_df[DS].apply(lambda feature: "RM" if feature in feature_groups["RM"] else 0)
    DS1_df["feature_group_LD"] = DS1_df[DS].apply(lambda feature: "LD" if feature in feature_groups["LD"] else 0)
    DS1_df["feature_group_G"] = DS1_df[DS].apply(lambda feature: "G" if feature in feature_groups["G"] else 0)
    DS1_df["feature_group_POS"] = DS1_df[DS].apply(lambda feature: "POS" if feature in feature_groups["POS"] else 0)
    DS1_df["feature_group_M"] = DS1_df[DS].apply(lambda feature: "M" if feature in feature_groups["M"] else 0)
    DS1_df["feature_group_C"] = DS1_df[DS].apply(lambda feature: "C" if feature in feature_groups["C"] else 0)
    DS1_df["feature_group_BC"] = DS1_df[DS].apply(lambda feature: "BC" if feature in feature_groups["BC"] else 0)
    DS1_df["feature_group_SDA"] = DS1_df[DS].apply(lambda feature: "SDA" if feature in feature_groups["SDA"] else 0)
    DS1_df["feature_group_SA"] = DS1_df[DS].apply(lambda feature: "SA" if feature in feature_groups["SA"] else 0)

    return DS1_df["feature_group_LS"].value_counts(), DS1_df["feature_group_RM"].value_counts(), DS1_df["feature_group_LD"].value_counts(), DS1_df["feature_group_G"].value_counts(), DS1_df["feature_group_POS"].value_counts(), DS1_df["feature_group_M"].value_counts(), DS1_df["feature_group_C"].value_counts(), DS1_df["feature_group_BC"].value_counts(), DS1_df["feature_group_SDA"].value_counts(), DS1_df["feature_group_SA"].value_counts()
```

```python
def get_groups_df(df, DS):

    DS1_df = pd.DataFrame(df[DS]).copy()
    DS1_df["feature_group_LS"] = DS1_df[DS].apply(lambda feature: "LS" if feature in feature_groups["LS"] else np.nan)
    DS1_df["feature_group_RM"] = DS1_df[DS].apply(lambda feature: "RM" if feature in feature_groups["RM"] else np.nan)
    DS1_df["feature_group_LD"] = DS1_df[DS].apply(lambda feature: "LD" if feature in feature_groups["LD"] else np.nan)
    DS1_df["feature_group_G"] = DS1_df[DS].apply(lambda feature: "G" if feature in feature_groups["G"] else np.nan)
    DS1_df["feature_group_POS"] = DS1_df[DS].apply(lambda feature: "POS" if feature in feature_groups["POS"] else np.nan)
    DS1_df["feature_group_M"] = DS1_df[DS].apply(lambda feature: "M" if feature in feature_groups["M"] else np.nan)
    DS1_df["feature_group_C"] = DS1_df[DS].apply(lambda feature: "C" if feature in feature_groups["C"] else np.nan)
    DS1_df["feature_group_BC"] = DS1_df[DS].apply(lambda feature: "BC" if feature in feature_groups["BC"] else np.nan)
    DS1_df["feature_group_SDA"] = DS1_df[DS].apply(lambda feature: "SDA" if feature in feature_groups["SDA"] else np.nan)
    DS1_df["feature_group_SA"] = DS1_df[DS].apply(lambda feature: "SA" if feature in feature_groups["SA"] else np.nan)
    DS1_df["feature_group"] = DS1_df.iloc[:, 1:].ffill(1).iloc[:,-1].copy()

    return DS1_df.iloc[:, [0, -1]]
    return DS1_df[["features", "feature_group"]]
```
